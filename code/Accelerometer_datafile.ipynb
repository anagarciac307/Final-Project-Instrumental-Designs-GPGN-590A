```python
#Working
from drawnow import *
import sys, os, serial, datetime
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import csv
%matplotlib


xlist = []
ylist = []
zlist = []

# Port needs to match that of arduino
ser = serial.Serial(port = 'COM14', baudrate=115200)
plt.ion()
count = 0


def makeFig():
    plt.ylim(-3,12)
    plt.title("Acceleration through time")
    plt.ylabel("acceleration (m/s^2)")
    plt.xlabel("Shot points")
    
    plt.plot(xlist, 'r-', label="x-direction")
    plt.plot(ylist, 'g-',label="y-direction")
    plt.plot(zlist, 'b-', label="z-direction")
    plt.legend()

with open ('accelerometer_baseline.csv', 'w', newline = '') as log:
    logWrite = csv.writer(log)
    while (True):

        line = ser.readline()
        lines = line.decode('utf-8')[:-2]
        print(lines)
        #lines = float(lines)
        lines = lines.split(',')
        xdir = float(lines[0])
        ydir = float(lines[1])
        zdir = float(lines[2])
        #print(xdir)
        #print(ydir)
        #print(zdir)
        xlist.append(xdir)
        ylist.append(ydir)
        zlist.append(zdir)



    #     xlist.append(lines)

        #print(xlist, ylist, zlist)
        

        drawnow(makeFig)

        plt.pause(0.001)

        count = count + 1
        if (count>1000):
            xlist.pop(0)
            ylist.pop(0)
            zlist.pop(0)
        
        time = count/10
        logWrite.writerow([xdir, ydir, zdir, time])
```
